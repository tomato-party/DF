```python
import pandas as pd
import numpy as np
import seaborn as sns

import scipy.stats as stats
from scipy.stats import pearsonr

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # to adjust the configuration of plot

pd.options.mode.chained_assignment = None
```

```python
df = pd.read_csv('movies.csv')
df
```

1. 작가와 제작 나라는 판단하지 않기 때문에 데이터 프레임에서 제거
2. 수익률이 타겟임으로 수익률 결측치 제거

```python
df = df.drop(['writer', 'country'], axis =1)
df = df.dropna(subset=['gross'])
df
```

# 수치데이터 

### 1. 평점에 따른 수익률

```python
filter_Nan_score_df = df.dropna(subset=['score'])
```

```python
filter_Nan_score_df.plot.scatter(x='score', y='gross')

plt.title('Score & Gross')
plt.xlabel('Score')
plt.ylabel('Gross')
plt.show()
```

```python
sns.regplot(x='score', y = 'gross', data =filter_Nan_score_df , scatter_kws={"color":"blue"}, line_kws={"color":"red"})
plt.show()
```

```python
corr, p_value = pearsonr(filter_Nan_score_df['gross'], filter_Nan_score_df['score'])
print("상관계수", corr)
print("p-value", p_value)
```

p-value > 0.05
=> 유의성 X

흥행률 관련 특성에서 제거

# 범주화 데이터

## 개봉날의 월로 새로운 데이터 칼럼 생성

```python
df['month'] = df['released'].str.split().str[0]
filter_Nan_month_df = df.dropna(subset=['month'])
filter_Nan_month_df
```

### 1. 개봉 월에 따른 수익률 평균

month데이터가 숫자가 아닌 달(영문) 데이터만 가져오기 

```python
filter_num = filter_Nan_month_df[~filter_Nan_month_df['month'].str.isdigit()]
filter_num_month_df = pd.DataFrame(filter_num)
filter_num_month_df['month'].unique()
```

```python
df_month_mean = pd.DataFrame(filter_num_month_df.groupby('month')['gross'].mean())
df_month_mean = df_month_mean.sort_values(by='gross', ascending = False)

df_month_mean
```

```python
plt.bar(df_month_mean.index, df_month_mean['gross'])
```

### 2. 계절에 따른 수익률 합

mapping 이용해 월별로 맞는 계절로 치환하기 

```python
mapping = {'spring': ['March','April','May'], 
           'summer':['June','July','August'], 
           'autumn':['September','October','November'],
           'winter':['December','January','February'] }
filter_num_month_df['month'] = filter_num_month_df['month'].map({m: season for season, months in mapping.items() for m in months})
filter_num_month_df
```

```python
df_season_mean = pd.DataFrame(filter_num_month_df.groupby('month')['gross'].mean())
df_season_mean = df_season_mean.sort_values(by='gross', ascending=False)

df_season_mean
```

```python
plt.bar(df_season_mean.index, df_season_mean['gross'])
```
